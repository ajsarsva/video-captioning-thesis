Empty Captions Verification

```python
for strategy in ['A', 'B', 'C']:
    path = f'/content/drive/MyDrive/thesis-data/captions/strategy_{strategy}.json'
    with open(path, 'r') as f:
        data = json.load(f)

    empty = [vid for vid, val in data.items() if not val['caption'].strip()]
    short = [vid for vid, val in data.items() if len(val['caption'].split()) < 3]

    print(f"Strategy {strategy}:")
    print(f"  Empty captions: {empty if empty else 'None ✅'}")
    print(f"  Very short captions (<3 words): {short if short else 'None ✅'}")
    print()
```
